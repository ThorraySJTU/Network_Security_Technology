```python
import math
def step(txt_file):
    with open(txt_file, "r") as f:
        data = f.readlines()
    step = []
    step_log = []
    for i in data:
        if i[:4]=='step':
            step.append(int(i.split(" = ")[1].split("\n")[0]))
    for i in range(len(step)):
        step_log.append(math.log(step[i],10))
    return step,step_log
```

```python
step_2,step_2_log = step('2.txt')
```

```python
step_3,step_3_log = step('3.txt')
```

```python
step_4,step_4_log = step('4.txt')
```

```python
print(step_2,step_3,step_4)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.boxplot([step_2_log,step_3_log,step_4_log],labels=['2','3','4'])
plt.title('Box graph of $log_{10}(steps)$ and n value')
plt.xlabel("n value")
plt.ylabel("$log_{10}(steps)$")
plt.show()
```
